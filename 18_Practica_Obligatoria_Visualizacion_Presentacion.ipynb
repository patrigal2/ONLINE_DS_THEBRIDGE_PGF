```python
df_titanic.info()
```

```python
colx = "alive"
coly = "who"
frec_relativa = df_titanic.groupby(coly, as_index = False)[colx].value_counts(normalize = True)
frec_relativa["proportion"] *= 100 

sns.catplot(x="alive",
            y = "proportion",
            col="who",
            hue = "alive",
            kind="bar", 
            edgecolor=".6",
            data=frec_relativa);
```

4- ... las diferencias de tasas de supervivencia entre mujeres y nios.

```python
plt.figure(figsize=(7, 5))
sns.countplot(x="alive", data=df_titanic, palette='viridis')
plt.title('Distribucin de la Supervivencia', fontsize=16)
plt.xlabel("Alive"(0 == no, 1 == yes), fontsize=12)
plt.ylabel("Total Pasajeros", fontsize=12)
plt.xticks([0, 1], ["no", "yes"])
plt.show();
```

```python
# Supervivencia segn sexo
plt.figure(figsize=(7, 5))
sns.countplot(x="who", hue="alive", data=df_titanic, palette='viridis')
plt.title('Supervivencia segn who', fontsize=16)
plt.xlabel("who", fontsize=12)
plt.ylabel("Total de pasajeros", fontsize=12)
plt.legend(title="alive", labels=['No', 'S'])
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(x="who", y="fare", data=df_titanic, hue="alive", palette="viridis", alpha=0.7)
plt.title("relacin entre who y fare", fontsize=16)
plt.xlabel("who", fontsize=12)
plt.ylabel("tarifa", fontsize=12)
plt.show()
```
