### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de una ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
#creamos un nuevo df con las ciudades que queremos mostrar en la grfica. 
ciudades_mas_pobladas = df_california["city"].isin(["LosAngeles", "SanDiego", "Sacramento", "Anaheim", "SantaAna"])
df_ciudades = df_california[ciudades_mas_pobladas]
df_ciudades = df_ciudades.sort_values(by="population_total", ascending=False)
df_ciudades
```

```python
#Creamos la grfica para mostrar la frecuencia absoluta

categorias = df_ciudades["city"].value_counts().index.to_list()
frecuencia_absoluta = df_ciudades["population_total"].to_list()

fig = plt.figure(figsize=(7,6)) 
ax = plt.axes() 

ax.bar(categorias, frecuencia_absoluta, color= ["blue","grey","red","olive", "brown"], alpha = 0.5)
ax.set_title("Ciudades ms pobladas de California")
ax.set_xlabel("Ciudades")
ax.set_ylabel("Habitantes (miles)")
ax.set_xticklabels(["Los Angeles", "San Diego", "Sacramento", "Anaheim", "Santa Ana"]);

#no se como poner los valores en cada barra. Entiendo que es con el bucle for pero...
```

```python
#HISTOGRAMA
#rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
data = df_california["area_total_km2"]

fig2 = plt.figure()
ax2 = plt.axes()
ax2.hist(data, bins=50, alpha=0.5, label="Hist Area")

font_dict = {"size": 18,
             "family": "Bradley Hand itc",
             "color": "steelblue"
}

ax2.set_xlabel("Superficie $(km^2)$", fontdict= font_dict)
ax2.set_ylabel("Numero de ciudades", fontdict= font_dict)
ax2.legend();
```

```python
#El scatter de rea de agua de una ciudad con su poblacion.

fig3 = plt.figure(figsize= (7,5))
ax3 = plt.axes()

pob, area = df_california["population_total"], df_california["area_water_km2"]

ax3.scatter(area, pob, c="green", alpha= 0.5)

ax3.set_ylabel("Poblacin(miles de habitantes)")
ax3.set_xlabel("Area de agua en la ciudad $(km^2)$")
ax3.set_title("Poblacin por kilmetro cuadrado de agua")

props_flecha = {
    "arrowstyle": "<->",
    "connectionstyle": "angle3"
} 

ax3.annotate("Outliers", xy= (80, 3.5), xytext=(90, 4.5),arrowprops= props_flecha);

#no se como poner el outliers ahi arriba
```

```python
#El boxplot de la altura de las ciudades.

data_elevation =df_california["elevation_m"]
```

```python
plt.boxplot(data_elevation, vert = False)
plt.xlabel("Altitud (m)")
plt.title("Diagrama de caja de la Altitud de las ciudades de California");
```
